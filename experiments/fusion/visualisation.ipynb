```julia
using Plots
using ColorSchemes

colors = ColorSchemes.tol_light

function gaussian(x, mean, height, width)
    return height * exp(-((x - mean) / width)^2)
end

x = range(0, 1, 150)
means = [0, 0.01, 0.1, 0.2, 0.4]
heights = [1, 1, 0.8, 0.8, 0.4]
widths = [0.02, 0.1, 0.1, 0.2, 0.1]

# Sum the profiles
sum_profile = zeros(length(x))
for i in 1:length(means)
    sum_profile += gaussian.(x, means[i], heights[i], widths[i])
end

# Plot
figure = plot()
for i in 1:length(means)
    plot!(x, gaussian.(x, means[i], heights[i], widths[i]), label="", z_order=:back, fill=true, fillalpha=0.8, linecolor=colors[i], fillcolor=colors[i])
end

# Plot labels separately to get nice legend order 
for i in 1:length(means)
    plot!([NaN], [NaN], label="Launcher $i", fill=true, fillalpha=0.8, linecolor=colors[i], fillcolor=colors[i])
end

plot!(x, sum_profile, linestyle=:dash, linecolor=:black, z_order=:front, label="Total")
xlabel!("Normalised radial coordinate")
ylabel!("Normalised power")
xlims!(0, 1)
ylims!(0, 2.8)
figure
```

```julia
savefig(figure, "fusion_profiles.png")
savefig(figure, "fusion_profiles.svg")
savefig(figure, "fusion_profiles.pdf")
```
